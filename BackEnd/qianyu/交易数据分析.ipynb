```python
data = pd.read_csv('data/ndm_transaction_list_all_20230705.csv', low_memory=False)
print(f'原数据列数：{len(data.columns)}\t\t\t\t原数据行数：{len(data)}')
data = data.dropna(axis=1, how='all')
print(f'删除空列后数据列数：{len(data.columns)}')
data = data.dropna(axis=1, thresh=len(data) * 0.8)
print(f'删除缺失项比例高于80%的列后数据列数：{len(data.columns)}\t数据行数：{len(data)}')

# 删除唯一值的列
unique_columns = [col for col in data.columns if data[col].nunique() == 1]
data = data.drop(columns=unique_columns)
print(f'删除唯一值的列后数据列数：{len(data.columns)}')
print(data.columns)
```

```python
# 按照交易次数划分不同债券分别统计交易情况
bond_grouped_data = data.groupby(['bond_cd', 'bnds_nm']).size().sort_values(ascending=False).reset_index(name='trade_count')
active_bonds = bond_grouped_data[bond_grouped_data['trade_count'] > 100]
print(f'活跃债数量：{len(active_bonds)}, 日交易>100')
mid_active_bonds = bond_grouped_data[(bond_grouped_data['trade_count'] <= 100) & (bond_grouped_data['trade_count'] > 10)]
print(f'中活跃债数量：{len(mid_active_bonds)}, 日交易10-100')
inactive_bonds = bond_grouped_data[(bond_grouped_data['trade_count'] <= 10) & (bond_grouped_data['trade_count'] > 1)]
print(f'非活跃债数量：{len(inactive_bonds)}, 日交易1-10')
silent_bonds = bond_grouped_data[bond_grouped_data['trade_count'] == 1]
print(f'不活跃债数量：{len(silent_bonds)}, 日交易=1')
```

```python
# 检查债券代码和债券名称的对应关系
code_name_mapping = data.groupby('bond_cd')['bnds_nm'].nunique()
name_code_mapping = data.groupby('bnds_nm')['bond_cd'].nunique()
if len(code_name_mapping[code_name_mapping > 1]) > 0:
    print("1. 债券代码对应多个名称的情况：")
    print(len(code_name_mapping[code_name_mapping > 1]))
elif len(name_code_mapping[name_code_mapping > 1]) > 0:
    print("\n2. 债券名称对应多个代码的情况：")
    print(len(name_code_mapping[name_code_mapping > 1]))
else:
    print("债券名称和债券代码是一一对应的")
```

```python
# 检查债券代码和到期收益率之间的对应关系
code_yld_mapping = data.groupby('bnds_nm')['yld_to_mrty'].nunique()
if len(code_yld_mapping[code_yld_mapping > 1]) > 0:
    print("存在一只债券对应多个到期收益率的情况：例如")
    print(code_yld_mapping[code_yld_mapping > 1].sort_values(ascending=False).head(2))
else:
    print("债券代码和到期收益率是一一对应的")
```

##### 交易与交易之间的关系

- 直接交易关系：A交易的卖方和B交易的买方是同一个机构，则A和B之间有直接关系。
- 交易传递方向：在上述直接交易关系中，如果A的时间比B晚（A在B之后发生），则边由B指向A。
可见[graph.html](Demo/交易关系特征/graph.html)
- 链路：价格约束，交易量约束，时间限制

存在的问题：
- 交易顺序：无法使用一天内的数据来做交易顺序的挖掘。收盘后统一结算当天所有交易，因此当天内的买卖顺序没有实际意义，可以先卖后买，单笔买和卖不决定收益，也可能不改变头寸。
- 交易量较大。在已经剔除当天对倒交易的情况下（5%），交易量仍然很大，如何仍然不能确定交易顺序，则交易关系网将很复杂。无法找出有价值的pattern。

```python
def remove_wash_trades(df):
    """
    对倒交易剔除函数
    """
    # 将时间列转换为datetime类型
    df['dl_tm'] = pd.to_datetime(df['dl_tm'])
    
    # 创建交易字典，用于快速查找
    trade_dict = {}
    wash_trade_codes = set()
    
    # 按债券代码和交易日期预处理数据
    for (bond_code, trade_date), group in df.groupby(['bond_cd', 'txn_dt']):
        # 按时间排序并创建关键特征元组
        trades = []
        for _, row in group.sort_values('dl_tm').iterrows():
            trade_info = {
                'dl_cd': row['dl_cd'],
                'dl_tm': row['dl_tm'],
                'nmnl_vol': row['nmnl_vol'],
                'net_prc': row['net_prc'],
                'buyer': row['byr_instn_cn_full_nm'],
                'buyer_trader': row['byr_trdr_nm'],
                'seller': row['slr_instn_cn_full_nm'],
                'seller_trader': row['slr_trdr_nm']
            }
            trades.append(trade_info)
        
        trade_dict[(bond_code, trade_date)] = trades
    
    # 查找对倒交易
    for trades in trade_dict.values():
        n = len(trades)
        matched = set()  # 使用集合存储已匹配的交易编号
        
        for i in range(n):
            if trades[i]['dl_cd'] in matched:
                continue
                
            # 使用二分查找优化时间范围内的交易查找
            current = trades[i]
            
            # 只需要检查后续交易
            for j in range(i + 1, n):
                if trades[j]['dl_cd'] in matched:
                    continue
                    
                next_trade = trades[j]
                
                # 快速检查匹配条件
                if (current['nmnl_vol'] == next_trade['nmnl_vol'] and 
                    current['net_prc'] == next_trade['net_prc'] and
                    ((current['buyer'] == next_trade['seller'] and current['seller'] == next_trade['buyer'] and current['buyer_trader'] == next_trade['seller_trader']) or
                    (current['buyer'] == next_trade['buyer'] and current['seller'] == next_trade['seller'] and current['buyer_trader'] == next_trade['buyer_trader']))):
                    
                    # 记录匹配的交易对
                    matched.add(current['dl_cd'])
                    matched.add(next_trade['dl_cd'])
                    wash_trade_codes.add(current['dl_cd'])
                    wash_trade_codes.add(next_trade['dl_cd'])
                    # 找到匹配后就跳出内层循环，确保每笔交易最多只被匹配一次
                    break
    
    # 使用交易编号过滤数据
    net_data = df[~df['dl_cd'].isin(wash_trade_codes)].copy()
    
    print(f"原始交易数量: {len(df)}")
    print(f"对倒交易数量: {len(wash_trade_codes)}")
    print(f"剔除后交易数量: {len(net_data)}")
    
    return net_data


print(f'一共有 {len(active_bonds_data_streamlined["net_prc"].unique())} 个不同的价格，\
      {len(active_bonds_data_streamlined["nmnl_vol"].unique())} 个不同的交易量')

net_data = remove_wash_trades(active_bonds_data_streamlined)
```
